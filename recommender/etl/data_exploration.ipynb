```python
import pandas as pd 
import numpy as np 
import json
```

```python
GR_BOOKS = '../data/goodreads_books.json'
GR_AUTHORS = '../data/goodreads_book_authors.json'
```

```python
books_list = []
limit = 100
i = 0
with open(GR_BOOKS) as f:
    for line in f:
        if i < limit:
            books_list.append(json.loads(line))
        i = i + 1

# df = pd.DataFrame(data)
# print(df.head(5))
print(books_list[0])
```

```python
df = pd.DataFrame(books_list)
df.head()
```

```python
df.drop(columns=['asin', 'kindle_asin', 'url'], inplace=True)
df.head()
```

```python
#make a string of title: {title} author: {author} series: {series} description: {descrioption} as a single string
#embed that string
#test it out

def build_string(row):
    return f"Title: {row['title']}\nDescription: {row['description']}"
```

```python
df['combined_string'] = df.apply(build_string, axis=1)
df.head()
```

```python
df.to_csv('../data/test.csv')
```
